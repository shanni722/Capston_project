```python
import yfinance as yf
import pandas as pd
from datetime import datetime

# Define a function to predict stock price for a given date
def predict_stock_price(stock_symbol, target_date):
    start = '2000-01-01'
    end = '2023-10-30'

    # Fetch historical stock data
    df = yf.download(stock_symbol, start, end)

    # Create a new DataFrame with only the 'Date' and 'Close' columns
    data = df[['Close']].reset_index()

    # Prepare the data for the model
    data['Date'] = pd.to_datetime(data['Date'])
    data['Date'] = data['Date'].map(pd.Timestamp.to_julian_date)
    X = data['Date'].values.reshape(-1, 1)
    y = data['Close']

    # Train a linear regression model
    from sklearn.linear_model import LinearRegression
    model = LinearRegression()
    model.fit(X, y)

    # Calculate the Julian date for the target date
    reference_date = datetime(1970, 1, 1)
    target_julian_date = (target_date - reference_date).days

    # Predict the stock price
    predicted_price = model.predict([[target_julian_date]])[0]

    return predicted_price

# Get user input for stock symbol and target date
stock_symbol = input("Enter a stock symbol (e.g., GOOGL, TSLA): ")
target_date_str = input("Enter the target date (YYYY-MM-DD): ")
target_date = datetime.strptime(target_date_str, '%Y-%m-%d')

# Predict stock price
predicted_price = predict_stock_price(stock_symbol, target_date)

print(f'Predicted price for {stock_symbol} on {target_date_str}: {predicted_price:.2f}')
```
